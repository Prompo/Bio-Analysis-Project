## Dataset Variables

**sex**<br>
0 Male<br>
1 Female<br>
**age_group**<br>
0  ages 18-24<br>
1  ages 25-34<br>
2  ages 35-44<br>
3  ages 45-54<br>
4  ages 55-64<br>
5  ages 65 and above<br>
**education**<br>
0 did not graduate High School<br>
1 Graduated High School<br>
2 Attended College or Technical School<br>
3 Graduated from College or Technical School<br>
**has_insurance**<br>
0 Do not have some form of health insurance<br>
1 Have some form of insurance<br>
**income_group**<br>
0 Less than 15,000<br>
1 15,000 to < 25,000<br>
2 25,000 to < 35,000<br>
3 35,000 to < 50,000<br>
4 50,000 to < 100,000<br>
5 100,000 to < 200,000<br>
6 200,000 or more<br>
**race**<br>
0 white<br>
1 Black only, non-Hispanic<br>
2 American Indian or Alaskan Native only, Non-Hispanic<br>
3 Asian only, non-Hispanic<br>
4 Native Hawaiian or other Pacific Islander only, Non-Hispanic<br>
5 Multiracial, non-Hispanic<br>
6 Hispanic<br>
**marital_status**<br>
0 married<br>
1 divorced<br>
2 widowed<br>
3 separated<br>
4 never married<br>
5 member of unmarried couple<br>
**general_health**<br>
0 excellent health<br>
1 Very good health<br>
2 good health<br>
3 fair health<br>
4 poor health<br>
**exercise_lately**: Exercise in Past 30 Days<br>
0 No<br>
1 Yes<br>
**sleep_time**<br>How Much Time Do You Sleep per day<br>
**smoking**: smoker status<br>
0 current smoker, everyday<br>
1 current smoker, occasionally<br>
2 former smoker<br>
3 never smoked<br>
**heavy_drinker**: Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week)<br>
**height**<br>
**weight**<br>
**bmi**<br>
**bmi_groups**: Four-categories of Body Mass Index (BMI)<br>
0 Underweight   (bmi<18.5)<br>
1 Normal weight (18.5<=bmi<25)<br>
2 Overweight    (25<=bmi<30)<br>
3 obese         (30<=bmi)<br>
**has_doctor**: Have Personal Health Care Provider?<br>
0 Yes, only one<br>
1 More than one<br>
2 No<br>
**affords_doctor**: Could Not Afford To See Doctor<br>
0 No<br>
1 Yes<br>
**last_checkup**: Length of time since last routine checkup<br>
0 Within past year<br>
1 Within past 2 years<br>
2 Within past 5 years<br>
3 5 or more years ago<br>
4 Never<br>
**heart_problem**: Having coronary heart disease (CHD) or myocardial infarction (MI)<br>
0 Did not report having MI or CHD<br>
1 Reported having MI or CHD<br>
**asthma_history**: asthma status<br>
0 Never<br>
1 Current<br>
2 Former<br>
**stroke**: Ever Diagnosed with a Stroke<br>
0 No<br>
1 Yes<br>
**depression**: Ever told you had a depressive disorder<br>
0 Yes<br>
1 No<br>
**blood_sugar**: When was your last blood test for high blood sugar?<br>
0 Within the past year<br>
1 Within the past 2 years<br>
2 Within the past 3 years<br>
3 Within the past 5 years<br>
4 Within the past 10 years<br>
5 10 or more years ago<br>
6 Never<br>
**prediabetes**:Ever been told by a doctor or other health professional that you have pre-diabetes or borderline diabetes?<br>
0 No, or yes during pregnancy<br>
1 Yes<br>
**diabetes_type**: What type of diabetes do you have?<br>
0, non diabetic<br>
1, type 1<br>
2, type 2<br>
**currently_insulin**: Now Taking Insulin<br>
0 No<br>
1 Yes<br>
**hemoglobin**: Times Checked for Glycosylated Hemoglobin in the past year<br>
**eye_exam**: Last Eye Exam Where Pupils Were Dilated<br>
0 Never<br>
1 Within the past month<br>
2 Within the past year<br>
3 Within the past 2 years<br>
4 2 or more years ago<br>
**eye_photo**:When was the last time a they took a photo of the back of your eye?<br>
0 Never<br>
1 Within the past month<br>
2 Within the past year<br>
3 Within the past 2 years<br>
4 2 or more years ago<br>
**diabetes_education**: When was the last time you took a course or class in how to manage your diabetes?<br>
0 Never<br>
1 Within the past year<br>
2 Within the last 2 years<br>
3 Within the last 3 years<br>
4 Within the last 5 years<br>
5 Within the last 10 years<br>
6 10 years ago or more<br>
**sore_feet** :Ever Had Feet Sores or Irritations Lasting More Than Four Weeks<br>
0 No<br>
1 Yes<br>
**diabetes**: (Ever told) you had diabetes<br>
0 No, or yes, but female told only during pregnancy<br>
1 pre-diabetes or borderline diabetes<br>
2 Yes<br>

```python
class create_dataset:
    def __init__(self, origin_dataset, columns_to_keep, column_values):
        self.data            = origin_dataset.copy()
        self.column_mapping  = columns_to_keep
        self.value_mapping   = column_values
        self.non_int_columns = [self.column_mapping['HTM4'],
                                self.column_mapping['WTKG3'],
                                self.column_mapping['_BMI5']]

    def filter_interviews(self):
        self.data = self.data[self.data['DISPCODE']==1100]

    def filter_columns(self):
        selected_cols = list(self.column_mapping.keys())
        self.data = self.data[selected_cols]

    def convert_values(self):
        for col,mapping in self.value_mapping.items():
            self.data[col] = self.data[col].map(lambda x: mapping.get(x, x))

    def filter_nan(self):
        self.data.dropna(inplace=True)

    def rename_columns(self):
        self.data.rename(columns=self.column_mapping, inplace=True)

    def convert_data_types(self):
        for col in list( set(self.data.columns) - set(self.non_int_columns) ):
            self.data[col] = self.data[col].astype(int)
        for col in self.non_int_columns:
            self.data[col] /= 100

    def convert_nan_to_zero(self):
        for col in ['PDIABTS1','PREDIAB2','DIABTYPE','INSULIN1','CHKHEMO3','EYEEXAM1','DIABEYE1','DIABEDU1','FEETSORE']:
            self.data[self.column_mapping[col]].fillna(0, inplace=True)

    def convert_nan_to_zero_2020(self):
        for col in ['PDIABTST','PREDIAB1','INSULIN1','CHKHEMO3','EYEEXAM1','DIABEYE','DIABEDU']:
            self.data[self.column_mapping[col]].fillna(0, inplace=True)

    def convert_nan_to_zero_2014(self):
        for col in ['PDIABTST','PREDIAB1','INSULIN','CHKHEMO3','EYEEXAM','DIABEYE','DIABEDU']:
            self.data[self.column_mapping[col]].fillna(0, inplace=True)

    def convert_nan_to_zero_2018(self):
        for col in ['PDIABTST','PREDIAB1','INSULIN','CHKHEMO3','EYEEXAM1','DIABEYE','DIABEDU']:
            self.data[self.column_mapping[col]].fillna(0, inplace=True)
```

```python
def dataset_creation_pipeline(inputs, column_names, column_values):
    obj = create_dataset(inputs, column_names, column_values)
    obj.filter_interviews()
    obj.filter_columns()
    obj.rename_columns()
    obj.convert_nan_to_zero_2018() ## !!! ###
    obj.convert_values()
    obj.filter_nan()
    obj.convert_data_types()
    return obj
```
